```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
df = pd.read_csv('diamond_data.csv', index_col = 0)
df2 = df.iloc[0:10000]
```

```python
df2.head()
```

```python
aa = ['color', 'clarity', 'cut']
for a in aa:
    plt.figure()
    graph = sns.kdeplot(
        data=df, 
        x='carat',
        hue = a, 
        multiple='stack',
        palette = 'crest',
        common_norm = False,
        linewidth = 0,
    )
    graph.set_title(a.upper())
```

```python
sns.histplot(
    data = df, 
    x = 'clarity',
    hue='color',
    palette = ["#0388FC", "#FC7703", "#65A857", "#9A57A8", "#C6396E", "#39C691", "#CB34A7"]
)
```

```python
g = sns.catplot(
    data=df, 
    col = 'clarity', 
    col_wrap = 4, 
    x = 'color', 
    kind='count', 
    height=2.5, 
    aspect=0.6, 
    palette = 'crest'
)
```

```python
sns.relplot(
    data=df2, 
    col = 'clarity',
    col_wrap = 4,
    x='carat', 
    y='price', 
    hue='color', 
    style = 'cut', 
    height = 5
)
```

```python
sns.lmplot(data=df[df.clarity == 'I1'], x='carat', y='price', hue='cut', height = 7.5)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.catplot(
    data=df[df.clarity == 'VVS1'],
    x='carat', 
    y = 'price',
    kind = 'violin'
)
```
